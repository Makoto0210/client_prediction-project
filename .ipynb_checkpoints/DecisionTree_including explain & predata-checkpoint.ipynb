```python
##資料說明 : 
    #目的:銀行用戶流失預測
    #1.RowNumber 編號
    #2.CustomerId 客戶id
    #3.Surname 姓氏
    #4.CreditScore 信用評分
    #5.Geography 客戶地區
    #6.Gender 性別
    #7.Age 年齡
    #8.Tenure 用戶時長
    #9.balance 存貸款情形
    #10.NumOfProducts 使用產品數量
    #11.HasCrCard 是否擁有信用卡
    #12.IsActiveMember 是否為活躍用戶
    #13.EstimatedSalary 估計收入
    #14.Exited 客戶是否已流失

#------------------------------------------------------------------------------------------------------------------------------------
#載入套件及讀取資料
import pandas as pd
from sklearn import tree
from sklearn.preprocessing import StandardScaler #標準化
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

**<font size=5 face=雅黑>dicition tree介紹</font>**

```python
#train資料選擇除了資料編號以及客戶id以外的資料做訓練並標準化並得出train_x ， train資料中exist作為train_y
train_x = train[['CreditScore', 'Geography', 'Gender', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'HasCrCard', 'IsActiveMember', 'EstimatedSalary']]
Std = StandardScaler()
train_x[['CreditScore', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'EstimatedSalary']] = Std.fit_transform(train_x[['CreditScore', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'EstimatedSalary']]) #標準化
train_x[['HasCrCard','IsActiveMember']] = train_x[['HasCrCard','IsActiveMember']].astype(str)
train_x_dummies = pd.get_dummies(train_x)
print(train_x_dummies)
train_y = train['Exited']
#------------------------------------------------------------------------------------------------------------------------------
test_x = test[['CreditScore', 'Geography', 'Gender', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'HasCrCard', 'IsActiveMember', 'EstimatedSalary']]
test_x[['CreditScore', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'EstimatedSalary']] = Std.fit_transform(test_x[['CreditScore', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'EstimatedSalary']]) #標準化
test_x[['HasCrCard','IsActiveMember']] = test_x[['HasCrCard','IsActiveMember']].astype(str)
test_x_dummies = pd.get_dummies(test_x)
print(test_x_dummies)

# 建立分類器
clf = tree.DecisionTreeClassifier(random_state=10)
exited_clf = clf.fit(train_x_dummies, train_y)
#---------------------------------------------
cclf = tree.DecisionTreeClassifier(criterion="entropy",random_state=10)
exited_cclf = cclf.fit(train_x_dummies, train_y)
# 預測
test_y_predicted = exited_clf.predict(test_x_dummies)
test_y_predicted1 = exited_cclf.predict(test_x_dummies)
predict = test_y_predicted.tolist()
predict1 = test_y_predicted1.tolist()
#-------------------------------------------------------
#把測試出來的test_y 加入test資料中最後一欄
test['Exited'] = predict
test['Exitedd'] = predict1
#-----------------------------------------------------------------
answer = {"RowNumber": test["RowNumber"],"Exited": test["Exited"]}
answer_df = pd.DataFrame(answer)
print(answer_df)
answer_df.to_csv("decisiontree-answer.csv")
#-----------------------------------------------------------------
answer1 = {"RowNumber": test["RowNumber"],"Exited": test["Exitedd"]}
answer1_df = pd.DataFrame(answer1)
answer1_df.to_csv("decisiontree-answer1.csv")

```

**<font size=5 face=雅黑>Precision, Recall, F1-score 意義</font>**

```python
TP(True Positive), 
TN(True Negative), 
FP(False Positive), 
FN(False Negative)

Recall(召回率) = TP/(TP+FN)
Precision(準確率) = TP/(TP+FP)
F1-score = 2 * Precision * Recall / (Precision + Recall)

-----------------------------------------------------------------------
Recall是在所有正樣本當中，能夠預測多少正樣本的比例，Precision為在所有預測為正樣本中，有多少為正樣本

一個Precision高而Recall低的模型跟一個Recall高而Precision低的模型，分別代表什麼意義呢？

前者，可以看作一個比較謹慎的模型，雖然常常沒辦法抓出命名實體，但只要有抓出幾乎都是正確的(Precision高)，而後者則是一個寬鬆的模型，雖然有時候會抓錯，但幾乎該抓的都有抓到(Recall高)

F1-score則是兩者的調和平均數，算是一個比較概略的指標來看這個模型的表現

--------------------------------------------------------------------------
如用Precision对系统进行评估，那么其回答的问题就是：

在诊断为癌症的一堆人中，到底有多少人真得了癌症？
如用Recall对系统进行评估，那么其回答的问题就是：

在一堆得了癌症的病人中，到底有多少人能被成功检测出癌症？
如用Accuracy对系统进行评估，那么其回答的问题就是：

在一堆癌症病人和正常人中，有多少人被系统给出了正确诊断结果（患癌或没患癌）？
```
