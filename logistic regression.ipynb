```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')

train_x = train[['CreditScore', 'Geography', 'Gender', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'HasCrCard', 'IsActiveMember', 'EstimatedSalary']]
Std = StandardScaler()
train_x[['CreditScore', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'EstimatedSalary']] = Std.fit_transform(train_x[['CreditScore', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'EstimatedSalary']]) #標準化
train_x[['HasCrCard','IsActiveMember']] = train_x[['HasCrCard','IsActiveMember']].astype(str)
train_x_dummies = pd.get_dummies(train_x)
print(train_x_dummies)
train_y = train['Exited']
#------------------------------------------------------------------------------------------------------------------------------
test_x = test[['CreditScore', 'Geography', 'Gender', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'HasCrCard', 'IsActiveMember', 'EstimatedSalary']]
test_x[['CreditScore', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'EstimatedSalary']] = Std.fit_transform(test_x[['CreditScore', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'EstimatedSalary']]) #標準化
test_x[['HasCrCard','IsActiveMember']] = test_x[['HasCrCard','IsActiveMember']].astype(str)
test_x_dummies = pd.get_dummies(test_x)
print(test_x_dummies)
#建立分類器-----------------------------
c = [0.01,1,100,10000,1000000]
#不做class_weight---------------------------------------------------------------------------------------
#-------------------------------------------------------------------------------------ridge
#建立分類器-----------------------------
model00 = LogisticRegression(C=c[0], penalty = "l2",random_state = 10) #ridge
model01 = LogisticRegression(C=c[1], penalty = "l2",random_state = 10)
model02 = LogisticRegression(C=c[2], penalty = "l2",random_state = 10)
model03 = LogisticRegression(C=c[3], penalty = "l2",random_state = 10)
model04 = LogisticRegression(C=c[4], penalty = "l2",random_state = 10)
#-------------------------------------------------------------------------
model00.fit(train_x_dummies,train_y)
test_y_predicted = model00.predict(test_x_dummies)
#-------------------------------------------------------------------------
predict = test_y_predicted.tolist()
#-------------------------------------------------------
test['EExited'] = predict
#-----------------------------------------------------------------
answer = {"RowNumber": test["RowNumber"],"Exited": test["EExited"]}
answer_df = pd.DataFrame(answer)
answer_df.to_csv("lrridgeanswerr.csv")
#--------------------------------------------------------------------------
model01.fit(train_x_dummies,train_y)
test_y_predicted = model01.predict(test_x_dummies)
#-------------------------------------------------------------------------
predict = test_y_predicted.tolist()
#-------------------------------------------------------
test['EExitedd'] = predict
#-----------------------------------------------------------------
answer = {"RowNumber": test["RowNumber"],"Exited": test["EExitedd"]}
answer_df = pd.DataFrame(answer)
answer_df.to_csv("lrridgeanswerr1.csv")
#-------------------------------------------------------------------------
model02.fit(train_x_dummies,train_y)
test_y_predicted = model02.predict(test_x_dummies)
#-------------------------------------------------------------------------
predict = test_y_predicted.tolist()
#-------------------------------------------------------
test['EExiteddd'] = predict
#-----------------------------------------------------------------
answer = {"RowNumber": test["RowNumber"],"Exited": test["EExiteddd"]}
answer_df = pd.DataFrame(answer)
answer_df.to_csv("lrridgeanswerr2.csv")
#------------------------------------------------------------------------
model03.fit(train_x_dummies,train_y)
test_y_predicted = model03.predict(test_x_dummies)
#-------------------------------------------------------------------------
predict = test_y_predicted.tolist()
#-------------------------------------------------------
test['EExitedddd'] = predict
#-----------------------------------------------------------------
answer = {"RowNumber": test["RowNumber"],"Exited": test["EExitedddd"]}
answer_df = pd.DataFrame(answer)
answer_df.to_csv("lrridgeanswerr3.csv")
#---------------------------------------------------------
model04.fit(train_x_dummies,train_y)
test_y_predicted = model04.predict(test_x_dummies)
#-------------------------------------------------------------------------
predict = test_y_predicted.tolist()
#-------------------------------------------------------
test['EExiteddddd'] = predict
#-----------------------------------------------------------------
answer = {"RowNumber": test["RowNumber"],"Exited": test["EExiteddddd"]}
answer_df = pd.DataFrame(answer)
answer_df.to_csv("lrridgeanswerr4.csv")
#---------------------------------------------------------
print(test[["EExited","EExitedd","EExiteddd","EExitedddd","EExiteddddd"]])
#-------------------------------------------------------------------------
acc = [0.7900,0.8250,0.8325,0.8300,0.8300]
preci = [0.4458,0.6000,0.6176,0.6061,0.6061]
fscore = [0.4684,0.3000,0.3853,0.3704,0.3704]
final = []
hw6 = []
i = 0
while i < len(acc):
    final.append(round(((acc[i] + preci[i] + fscore[i])/3),4))
    hw6.append(round((0.3 * acc[i] + 0.3 * preci[i] + 0.4 * fscore[i]),4))
    i = i + 1
print(final)
print(hw6)
answer = {"Accuracy": acc, "Precision": preci, "F-score": fscore, "Final": final, "HW6-Grade": hw6}
answer_df = pd.DataFrame(answer)
print(answer_df)
answer_df.to_csv("logisticreg-result.csv")
```
